```python
#1
def RemoveRowCol(A, K, L):
    M = len(A)
    N = len(A[0])
    
    # Преобразуем в индексы с 0
    K = K - 1
    L = L - 1
    
    if K >= M or L >= N or K < 0 or L < 0:
        return A
    
    new_A = []
    for i in range(M):
        if i != K:
            row = []
            for j in range(N):
                if j != L:
                    row.append(A[i][j])
            new_A.append(row)
    return new_A

M, N = map(int, input("M N: ").split())
A = [list(map(int, input().split())) for _ in range(M)]
K, L = map(int, input("K L: ").split())
result = RemoveRowCol(A, K, L)
for row in result:
    print(row)
```

```python
#2
def Split(A):
    B = [x for x in A if x % 2 == 0]
    C = [x for x in A if x % 2 != 0]
    return B, C

A = list(map(int, input("Список: ").split()))
B, C = Split(A)
print("Четные:", B)
print("Нечетные:", C)
```

```python
#3
def SwapRow(A, K1, K2):
    A[K1], A[K2] = A[K2], A[K1]
    return A

M, N = map(int, input("M N: ").split())
A = [list(map(float, input().split())) for _ in range(M)]
K1, K2 = map(int, input("K1 K2: ").split())
result = SwapRow(A, K1, K2)
for row in result:
    print(row)
```

```python
#4
def Transp(A):
    M = len(A)
    return [[A[j][i] for j in range(M)] for i in range(M)]

M = int(input("M: "))
A = [list(map(float, input().split())) for _ in range(M)]
result = Transp(A)
for row in result:
    print(row)
```

```python
#5
def MtrProd(A, B):
    rows_A = len(A)
    cols_A = len(A[0])
    cols_B = len(B[0])
    
    result = [[0 for _ in range(cols_B)] for _ in range(rows_A)]
    
    for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                result[i][j] += A[i][k] * B[k][j]
    return result

# Матрица A
M, N = map(int, input("A: M N: ").split())
A = [list(map(int, input().split())) for _ in range(M)]

# Матрица B  
N2, P = map(int, input("B: N P: ").split())
B = [list(map(int, input().split())) for _ in range(N2)]

if N != N2:
    print("Несовместимые размеры")
else:
    result = MtrProd(A, B)
    for row in result:
        print(row)
```

```python
#6
def SumMtr(A, B):
    M = len(A)
    N = len(A[0])
    result = [[A[i][j] + B[i][j] for j in range(N)] for i in range(M)]
    return result

M, N = map(int, input("M N: ").split())
A = [list(map(int, input().split())) for _ in range(M)]
B = [list(map(int, input().split())) for _ in range(M)]
result = SumMtr(A, B)
for row in result:
    print(row)
```

```python
#7
def SumDiagonal(A, main=True):
    n = len(A)
    total = 0
    for i in range(n):
        if main:
            total += A[i][i]
        else:
            total += A[i][n-1-i]
    return total

n = int(input("n: "))
A = [list(map(int, input().split())) for _ in range(n)]
choice = input("Главная (g) или побочная (p): ")
result = SumDiagonal(A, choice.lower() == 'g')
print("Сумма:", result)
```

```python
#8
def RemoveRows(A, K1, K2):
    M = len(A)
    if K1 > M:
        return A
    
    K2 = min(K2, M-1)
    return [A[i] for i in range(M) if i < K1 or i > K2]

M, N = map(int, input("M N: ").split())
A = [list(map(int, input().split())) for _ in range(M)]
K1, K2 = map(int, input("K1 K2: ").split())
result = RemoveRows(A, K1, K2)
for row in result:
    print(row)
```

```python
#9
def InvertStr(S, K, N):
    if K >= len(S):
        return ""
    end = min(K + N, len(S))
    substring = S[K:end]
    return substring[::-1]

S = input("Строка S: ")
K = int(input("K: "))
N = int(input("N: "))
result = InvertStr(S, K, N)
print("Результат:", result)
```

```python
#10
def SwapCol(A, K1, K2):
    for i in range(len(A)):
        A[i][K1], A[i][K2] = A[i][K2], A[i][K1]
    return A

M, N = map(int, input("M N: ").split())
A = [list(map(int, input().split())) for _ in range(M)]
K1, K2 = map(int, input("K1 K2: ").split())
result = SwapCol(A, K1, K2)
for row in result:
    print(row)
```

```python
#11
def CompressStr(S):
    result = []
    i = 0
    while i < len(S):
        count = 1
        while i + count < len(S) and S[i + count] == S[i]:
            count += 1
        if count > 4:
            result.append(S[i] + '[' + str(count) + ']')
        else:
            result.append(S[i] * count)
        i += count
    return ''.join(result)

S = input("Строка S: ")
result = CompressStr(S)
print("Сжатая строка:", result)
```

```python
#12
def IsIdent(s):
    if not s or s[0].isdigit():
        return False
    for char in s:
        if not (char.isalnum() or char == '_'):
            return False
    return True

s = input("Строка: ")
print("Допустимый идентификатор:" if IsIdent(s) else "Не допустимый идентификатор")
```

```python
#13
def Chessboard(M, N):
    A = [[(i + j) % 2 for j in range(N)] for i in range(M)]
    return A

M = int(input("M: "))
N = int(input("N: "))
result = Chessboard(M, N)
for row in result:
    print(row)
```

```python
#14
def Bell(L):
    L_sorted = sorted(L)
    result = []
    left = True
    while L_sorted:
        if left:
            result.append(L_sorted.pop(0))
        else:
            result.append(L_sorted.pop())
        left = not left
    return result

L = list(map(int, input("Список: ").split()))
result = Bell(L)
print("Колокол:", result)
```
