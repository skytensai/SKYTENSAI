```python
#1
M, N = map(int, input("M N: ").split())
matrix = [list(map(int, input().split())) for _ in range(M)]

found = 0
for j in range(N):
    all_odd = True
    for i in range(M):
        if matrix[i][j] % 2 == 0:
            all_odd = False
            break
    if all_odd:
        found = j + 1
        break

print(found)
```

```python
#2
M, N = map(int, input("M N: ").split())
matrix = [list(map(int, input().split())) for _ in range(M)]

for i in range(M):
    min_idx = matrix[i].index(min(matrix[i]))
    max_idx = matrix[i].index(max(matrix[i]))
    matrix[i][min_idx], matrix[i][max_idx] = matrix[i][max_idx], matrix[i][min_idx]

for row in matrix:
    print(*row)
```

```python
#3
M, N = map(int, input("M N: ").split())
matrix = [list(map(int, input().split())) for _ in range(M)]

min_val = min(min(row) for row in matrix)
max_val = max(max(row) for row in matrix)

min_row = max_row = -1
for i in range(M):
    if min_val in matrix[i]:
        min_row = i
    if max_val in matrix[i]:
        max_row = i

if min_row != max_row:
    matrix[min_row], matrix[max_row] = matrix[max_row], matrix[min_row]

for row in matrix:
    print(*row)
```

```python
#6
M, N = map(int, input("M N: ").split())
matrix = [list(map(int, input().split())) for _ in range(M)]

max_val = max(max(row) for row in matrix)
max_col = -1
for j in range(N):
    for i in range(M):
        if matrix[i][j] == max_val:
            max_col = j
            break
    if max_col != -1:
        break

for i in range(M):
    del matrix[i][max_col]

for row in matrix:
    print(*row)
```

```python
#8
M, N = map(int, input("M N: ").split())
matrix = [list(map(int, input().split())) for _ in range(M)]

max_val = max(max(row) for row in matrix)
max_row = -1
for i in range(M):
    if max_val in matrix[i]:
        max_row = i
        break

matrix.insert(max_row, matrix[max_row][:])

for row in matrix:
    print(*row)
```

```python
#9
M, N = map(int, input("M N: ").split())
matrix = [list(map(int, input().split())) for _ in range(M)]

for j in range(N):
    col_sum = sum(matrix[i][j] for i in range(M))
    average = col_sum / M
    count = sum(1 for i in range(M) if matrix[i][j] > average)
    print(f"Столбец {j+1}: {count}")
```

```python
#10
M, N = map(int, input("M N: ").split())
matrix = [list(map(int, input().split())) for _ in range(M)]

half_M, half_N = M // 2, N // 2
for i in range(half_M):
    for j in range(half_N):
        matrix[i][j], matrix[i + half_M][j + half_N] = matrix[i + half_M][j + half_N], matrix[i][j]

for row in matrix:
    print(*row)
```

```python
#12
M, N = map(int, input("M N: ").split())
matrix = [list(map(int, input().split())) for _ in range(M)]

max_sum = -1
max_row = -1
for i in range(M):
    row_sum = sum(matrix[i])
    if row_sum > max_sum:
        max_sum = row_sum
        max_row = i + 1

print(f"Строка: {max_row}, Сумма: {max_sum}")
```

```python
#13
M, N = map(int, input("M N: ").split())
matrix = [list(map(int, input().split())) for _ in range(M)]

min_val = min(min(row) for row in matrix)
min_row = -1
for i in range(M):
    if min_val in matrix[i]:
        min_row = i
        break

matrix.insert(min_row, [0] * N)

for row in matrix:
    print(*row)
```

```python
#14
M, N = map(int, input("M N: ").split())
matrix = [list(map(int, input().split())) for _ in range(M)]

last_even_row = 0
for i in range(M):
    all_even = True
    for j in range(N):
        if matrix[i][j] % 2 != 0:
            all_even = False
            break
    if all_even:
        last_even_row = i + 1

print(last_even_row)
```

```python
#15
M, N = map(int, input("M N: ").split())
matrix = [list(map(int, input().split())) for _ in range(M)]

last_positive_col = -1
for j in range(N):
    all_positive = True
    for i in range(M):
        if matrix[i][j] <= 0:
            all_positive = False
            break
    if all_positive:
        last_positive_col = j

if last_positive_col != -1 and last_positive_col != 0:
    for i in range(M):
        matrix[i][0], matrix[i][last_positive_col] = matrix[i][last_positive_col], matrix[i][0]

for row in matrix:
    print(*row)
```
