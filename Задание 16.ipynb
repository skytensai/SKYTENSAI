```python
# 1а) Функция full_name
def full_name(first_name, last_name, middle_name=None):
    if middle_name:
        return f"{last_name} {first_name} {middle_name}"
    else:
        return f"{last_name} {first_name}"

# Проверка работы функции
print(full_name('Иван', 'Петров'))
print(full_name('Мария', 'Сидорова', middle_name='Алексеевна'))
```

```python
# 1б) Функция transform_list
def transform_list(numbers, transform_function=None):
    if transform_function:
        return [transform_function(x) for x in numbers]
    else:
        return numbers.copy()

# Проверка работы функции
print(transform_list([1, 2, 3]))
print(transform_list([1, 2, 3], transform_function=lambda x: x * 2))
```

```python
# 1в) Функция sum_positive_numbers
def sum_positive_numbers(*args):
    return sum(x for x in args if x > 0)

# Проверка работы функции
print(sum_positive_numbers(1, -2, 3, -4, 5))
print(sum_positive_numbers(-1, -2, -3))
```

```python
# 1г) Функция common_keys
def common_keys(**kwargs):
    dicts = list(kwargs.values())
    if not dicts:
        return {}
    
    common = set(dicts[0].keys())
    for d in dicts[1:]:
        common = common.intersection(set(d.keys()))
    
    return {key: dicts[-1][key] for key in common}

# Проверка работы функции
print(common_keys(dict1={'a': 1, 'b': 2}, dict2={'b': 2, 'c': 3}))
print(common_keys(dict1={'x': 5, 'y': 6}, dict2={'y': 6, 'z': 7}, dict3={'y': 6, 'w': 8}))
```

```python
# 2в) Функция multiply_all
def multiply_all(*args):
    result = 1
    for num in args:
        result *= num
    return result

# Проверка работы функции
print(multiply_all(1, 2, 3, 4))
print(multiply_all(5, 10))
```

```python
# 2г) Функция filter_strings
def filter_strings(strings, **criteria):
    result = []
    for string in strings:
        valid = True
        if 'min_length' in criteria and len(string) < criteria['min_length']:
            valid = False
        if 'starts_with' in criteria and not string.startswith(criteria['starts_with']):
            valid = False
        if 'ends_with' in criteria and not string.endswith(criteria['ends_with']):
            valid = False
        
        if valid:
            result.append(string)
    return result

# Проверка работы функции
print(filter_strings(['яблоко', 'вишня', 'киви'], min_length=6))
print(filter_strings(['test', 'text', 'tent'], starts_with='te', ends_with='t'))
```

```python
# 3б) Функция compare_elements
def compare_elements(list1, list2, comparison_function=None):
    if comparison_function is None:
        comparison_function = lambda x, y: x == y
    
    result = []
    for i in range(min(len(list1), len(list2))):
        if comparison_function(list1[i], list2[i]):
            result.append((list1[i], list2[i]))
    return result

# Проверка работы функции
print(compare_elements([1, 2, 3], [1, 4, 3]))
print(compare_elements([1, 2, 3], [4, 5, 6], comparison_function=lambda x, y: x + y > 5))
```

```python
# 4а) Функция calculate
def calculate(a, b, operation='add'):
    operations = {
        'add': a + b,
        'subtract': a - b,
        'multiply': a * b,
        'divide': a / b
    }
    return operations.get(operation, 'Неизвестная операция')

# Проверка работы функции
print(calculate(10, 5))
print(calculate(10, 5, operation='multiply'))
```

```python
# 4б) Функция modify_strings
def modify_strings(strings, case=None):
    if case == 'upper':
        return [s.upper() for s in strings]
    elif case == 'lower':
        return [s.lower() for s in strings]
    else:
        return strings

# Проверка работы функции
print(modify_strings(['Hello', 'World']))
print(modify_strings(['Hello', 'World'], case='upper'))
```

```python
# 4в) Функция average
def average(*args):
    if not args:
        return 0
    return sum(args) / len(args)

# Проверка работы функции
print(average(1, 2, 3, 4, 5))
print(average(10, 20))
```

```python
# 5г) Функция find_primes
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def find_primes(*numbers):
    return [num for num in numbers if is_prime(num)]

# Проверка работы функции
print(find_primes(2, 3, 4, 5, 6, 7))
print(find_primes(10, 15, 17, 19, 23))
```

```python
# 6а) Функция word_count
def word_count(text, separator=' '):
    words = text.split(separator)
    count_dict = {}
    for word in words:
        count_dict[word] = count_dict.get(word, 0) + 1
    return count_dict

# Проверка работы функции
print(word_count('один два один'))
print(word_count('red,green,blue,red', separator=','))
```

```python
# 6в) Функция merge_and_sort_keys
def merge_and_sort_keys(*dicts):
    merged = {}
    for d in dicts:
        for key, value in d.items():
            if key not in merged or value > merged[key]:
                merged[key] = value
    
    sorted_keys = sorted(merged.keys(), key=lambda k: merged[k], reverse=True)
    return sorted_keys

# Проверка работы функции
print(merge_and_sort_keys({'a': 2, 'b': 3}, {'a': 5, 'c': 1}))
print(merge_and_sort_keys({'x': 10}, {'y': 5}, {'z': 15}))
```

```python
# 9г) Функция unique_keys
def unique_keys(*dicts):
    all_keys = {}
    for d in dicts:
        for key, value in d.items():
            if key not in all_keys:
                all_keys[key] = value
            else:
                all_keys[key] = None
    
    return {key: value for key, value in all_keys.items() if value is not None}

# Проверка работы функции
print(unique_keys({'a': 1, 'b': 2}, {'b': 3, 'c': 4}))
print(unique_keys({'x': 5}, {'y': 6}, {'z': 7}, {'x': 8}))
```

```python
# 10г) Функция product_non_zero
def product_non_zero(*args):
    result = 1
    has_non_zero = False
    for num in args:
        if num != 0:
            result *= num
            has_non_zero = True
    return result if has_non_zero else 1

# Проверка работы функции
print(product_non_zero(1, 2, 0, 4))
print(product_non_zero(0, 0, 0))
```

```python
# 12б) Функция process_string
def process_string(text, separator=None):
    if separator:
        words = text.split(separator)
        capitalized_words = [word.capitalize() for word in words]
        return separator.join(capitalized_words)
    else:
        return text.lower()

# Проверка работы функции
print(process_string('HELLO WORLD'))
print(process_string('python_java_c++', separator='_'))
```

```python
# 12в) Функция difference
def difference(first_list, *other_lists):
    result = []
    for item in first_list:
        in_all_others = True
        for other_list in other_lists:
            if item not in other_list:
                in_all_others = False
                break
        if not in_all_others:
            result.append(item)
    return result

# Проверка работы функции
print(difference([1, 2, 3], [2, 3, 4], [3, 4, 5]))
print(difference(['a', 'b', 'c'], ['b', 'c'], ['c']))
```

```python
# 12г) Функция common_in_at_least_two
def common_in_at_least_two(*lists):
    from collections import defaultdict
    count_dict = defaultdict(int)
    
    for lst in lists:
        unique_items = set(lst)
        for item in unique_items:
            count_dict[item] += 1
    
    return [item for item, count in count_dict.items() if count >= 2]

# Проверка работы функции
print(common_in_at_least_two([1, 2], [2, 3], [2, 4]))
print(common_in_at_least_two(['a', 'b'], ['b', 'c'], ['b', 'd']))
```

```python
# 13в) Функция sorted_values_by_keys
def sorted_values_by_keys(*dicts):
    all_items = []
    for d in dicts:
        all_items.extend(d.items())
    
    sorted_items = sorted(all_items, key=lambda x: int(x[0]) if x[0].isdigit() else x[0])
    return [value for key, value in sorted_items]

# Проверка работы функции
print(sorted_values_by_keys({'2': 'b', '1': 'a'}, {'4': 'd', '3': 'c'}))
print(sorted_values_by_keys({'10': 'j'}, {'5': 'e', '15': 'o'}))
```

```python
# 14б) Функция extract_digits
def extract_digits(strings, digits_only=False):
    if digits_only:
        result = []
        for s in strings:
            digits = ''.join(char for char in s if char.isdigit())
            result.append(digits)
        return result
    else:
        return strings

# Проверка работы функции
print(extract_digits(['a1b2', 'c3d4']))
print(extract_digits(['a1b2', 'c3d4'], digits_only=True))
```
