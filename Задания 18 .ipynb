```python
def segment_midpoint(x1, y1, x2, y2):
    return (x1 + x2) / 2, (y1 + y2) / 2

def can_form_triangle(a, b, c):
    return a + b > c and a + c > b and b + c > a

def triangle_area(a, b, c):
    p = (a + b + c) / 2
    return (p * (p - a) * (p - b) * (p - c)) ** 0.5
```

```python
import math

def distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

def triangle_perimeter(x1, y1, x2, y2, x3, y3):
    a = distance(x1, y1, x2, y2)
    b = distance(x2, y2, x3, y3)
    c = distance(x1, y1, x3, y3)
    return a + b + c if a + b > c and a + c > b and b + c > a else None

def triangle_area_points(x1, y1, x2, y2, x3, y3):
    a = distance(x1, y1, x2, y2)
    b = distance(x2, y2, x3, y3)
    c = distance(x1, y1, x3, y3)
    if not (a + b > c and a + c > b and b + c > a):
        return None
    p = (a + b + c) / 2
    return math.sqrt(p * (p - a) * (p - b) * (p - c))
```

```python
import math

def gcd(a, b):
    return math.gcd(abs(a), b)

def simplify_fraction(p, q):
    d = gcd(p, q)
    return p // d, q // d

def fraction_to_decimal(p, q):
    return p / q
```

```python
def line_equation(x1, y1, x2, y2):
    if x1 == x2:
        return None, x1
    k = (y2 - y1) / (x2 - x1)
    b = y1 - k * x1
    return k, b

def are_parallel(k1, b1, k2, b2):
    return k1 == k2

def point_on_line(k, b, x, y):
    if k is None:
        return x == b
    return abs(y - (k * x + b)) < 1e-10
```

```python
import math

def circle_area(r):
    return math.pi * r ** 2

def circle_length(r):
    return 2 * math.pi * r

def sector_area(r, angle):
    return math.pi * r ** 2 * angle / 360

def inscribed_square_perimeter(r):
    side = r * math.sqrt(2)
    return 4 * side
```

```python
def remove_char(s, char):
    return s.replace(char, '')

def replace_char(s, old, new):
    return s.replace(old, new)

def count_char(s, char):
    return s.count(char)
```

```python
def print_matrix(matrix):
    for row in matrix:
        print(row)

def add_matrices(a, b):
    return [[a[i][j] + b[i][j] for j in range(len(a[0]))] for i in range(len(a))]

def multiply_matrices(a, b):
    result = [[0 for _ in range(len(b[0]))] for _ in range(len(a))]
    for i in range(len(a)):
        for j in range(len(b[0])):
            for k in range(len(b)):
                result[i][j] += a[i][k] * b[k][j]
    return result

def transpose_matrix(matrix):
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
```

```python
def vector_add(v1, v2):
    return [v1[i] + v2[i] for i in range(len(v1))]

def vector_multiply_scalar(v, scalar):
    return [x * scalar for x in v]

def vector_min_max(v):
    return min(v), max(v)
```

```python
def remove_digits(filename):
    with open(filename, 'r') as f:
        content = f.read()
    result = ''.join([c for c in content if not c.isdigit()])
    with open(filename, 'w') as f:
        f.write(result)

def count_chars(filename):
    with open(filename, 'r') as f:
        return len(f.read())

def count_lines(filename):
    with open(filename, 'r') as f:
        return len(f.readlines())
```

```python
def complex_abs(z):
    return (z.real ** 2 + z.imag ** 2) ** 0.5

def complex_multiply(z1, z2):
    return complex(z1.real * z2.real - z1.imag * z2.imag,
                   z1.real * z2.imag + z1.imag * z2.real)

def complex_divide(z1, z2):
    denominator = z2.real ** 2 + z2.imag ** 2
    return complex((z1.real * z2.real + z1.imag * z2.imag) / denominator,
                   (z1.imag * z2.real - z1.real * z2.imag) / denominator)
```

```python
def power_fraction(p, q, n):
    return p ** n, q ** n

def fractions_equal(p1, q1, p2, q2):
    return p1 * q2 == p2 * q1

def fraction_greater(p1, q1, p2, q2):
    return p1 * q2 > p2 * q1

def fraction_less(p1, q1, p2, q2):
    return p1 * q2 < p2 * q1
```

```python
import math

def vector_subtract(v1, v2):
    return (v1[0] - v2[0], v1[1] - v2[1])

def dot_product(v1, v2):
    return v1[0] * v2[0] + v1[1] * v2[1]

def vector_length(v):
    return math.sqrt(v[0] ** 2 + v[1] ** 2)
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```

```python
def matrix_min(matrix):
    return min(min(row) for row in matrix)

def matrix_max(matrix):
    return max(max(row) for row in matrix)

def matrix_sum(matrix):
    return sum(sum(row) for row in matrix)

def row_sum(matrix, row_index):
    return sum(matrix[row_index])
```
