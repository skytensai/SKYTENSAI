```python
# Ввод данных
a = float(input("Введите длину a: "))
b = float(input("Введите ширину b: "))
c = float(input("Введите высоту c: "))

# Вычисления
V = a * b * c
S = 2 * (a*b + b*c + a*c)

# Вывод результатов
print("Объем:", V)
print("Площадь поверхности:", S)
```

```python
# Ввод данных
R = float(input("Введите радиус R: "))

# Вычисления
import math
L = 2 * math.pi * R
S = math.pi * R ** 2

# Вывод результатов
print("Длина окружности:", L)
print("Площадь круга:", S)
```

```python
# Ввод данных
R1 = float(input("Введите внешний радиус R1: "))
R2 = float(input("Введите внутренний радиус R2: "))

# Вычисления
import math
S1 = math.pi * R1**2
S2 = math.pi * R2**2
S3 = S1 - S2

# Вывод результатов
print("Площадь большого круга:", S1)
print("Площадь малого круга:", S2)
print("Площадь кольца:", S3)
```

```python
# Ввод данных
x1 = float(input("Введите x1: "))
y1 = float(input("Введите y1: "))
x2 = float(input("Введите x2: "))
y2 = float(input("Введите y2: "))

# Вычисления
width = abs(x2 - x1)
height = abs(y2 - y1)
P = 2 * (width + height)
S = width * height

# Вывод результатов
print("Периметр:", P)
print("Площадь:", S)
```

```python
import math

x1 = float(input("x1: "))
y1 = float(input("y1: "))
x2 = float(input("x2: "))
y2 = float(input("y2: "))
x3 = float(input("x3: "))
y3 = float(input("y3: "))

a = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
b = math.sqrt((x3 - x2)**2 + (y3 - y2)**2)
c = math.sqrt((x1 - x3)**2 + (y1 - y3)**2)

perimeter = a + b + c
p = perimeter / 2
area = math.sqrt(p * (p - a) * (p - b) * (p - c))

print("Периметр:", perimeter)
print("Площадь:", area)
```

```python
X = float(input("Кг шоколадных конфет: "))
A = float(input("Стоимость шоколадных: "))
Y = float(input("Кг ирисок: "))
B = float(input("Стоимость ирисок: "))

price_choc = A / X
price_iris = B / Y
ratio = price_choc / price_iris

print("1 кг шоколадных:", price_choc)
print("1 кг ирисок:", price_iris)
print("Во сколько раз дороже:", ratio)
```
