```python
# 1) Функции Fact и Fact2 для вычисления факториала и двойного факториала
def Fact(N):
    if N == 0 or N == 1:
        return 1
    else:
        return N * Fact(N - 1)

def Fact2(N):
    if N == 1 or N == 2:
        return N
    else:
        return N * Fact2(N - 2)

print("Факториал 5:", Fact(5))
print("Двойной факториал 7:", Fact2(7))
```

```python
# 2) Функция PowerN для вычисления степени числа
def PowerN(x, n):
    if n == 0:
        return 1
    elif n > 0:
        return x * PowerN(x, n - 1)
    else:
        return 1 / PowerN(x, -n)

print("2^3 =", PowerN(2, 3))
print("2^(-3) =", PowerN(2, -3))
```

```python
# 3) Функция SqrtK для приближенного вычисления корня k-й степени
def SqrtK(x, k, n):
    if n == 0:
        return 1
    else:
        y_prev = SqrtK(x, k, n - 1)
        return y_prev - (y_prev - x / (y_prev ** (k - 1))) / k

print("Корень 3-й степени из 8 (2 итерации):", SqrtK(8, 3, 2))
print("Корень 4-й степени из 16 (3 итерации):", SqrtK(16, 4, 3))
```

```python
# 4) Функция FibRec для вычисления чисел Фибоначчи с подсчетом вызовов
call_count = 0

def FibRec(N):
    global call_count
    call_count += 1
    
    if N == 1 or N == 2:
        return 1
    else:
        return FibRec(N - 2) + FibRec(N - 1)

# Находим 5 чисел Фибоначчи
fib_numbers = [3, 5, 7, 10, 12]
results = []

for num in fib_numbers:
    call_count = 0
    fib_value = FibRec(num)
    results.append((num, fib_value, call_count))

print("Числа Фибоначчи:")
for n, fib, calls in results:
    print(f"F({n}) = {fib}, рекурсивных вызовов: {calls}")
```

```python
# 5) Функция C для вычисления числа сочетаний с подсчетом вызовов
c_call_count = 0

def C(m, n):
    global c_call_count
    c_call_count += 1
    
    if m == 0 or m == n:
        return 1
    else:
        return C(m, n - 1) + C(m - 1, n - 1)

# Вычисляем сочетания для разных m при фиксированном n
N = 6
M_values = [1, 2, 3, 4, 5]
results = []

for m in M_values:
    c_call_count = 0
    comb_value = C(m, N)
    results.append((m, N, comb_value, c_call_count))

print("Числа сочетаний:")
for m, n, comb, calls in results:
    print(f"C({m}, {n}) = {comb}, рекурсивных вызовов: {calls}")
```

```python
# 6) Функция NOD для нахождения наибольшего общего делителя
def NOD(A, B):
    if A == 0:
        return B
    else:
        return NOD(B % A, A)

# Находим НОД для пар чисел
A, B, C, D = 36, 48, 54, 24

print(f"НОД({A}, {B}) = {NOD(A, B)}")
print(f"НОД({A}, {C}) = {NOD(A, C)}")
print(f"НОД({A}, {D}) = {NOD(A, D)}")
```

```python
# 7) Функция MinRec для нахождения минимального элемента в массиве
def MinRec(A, N):
    if N == 1:
        return A[0]
    else:
        return min(A[N - 1], MinRec(A, N - 1))

# Тестируем на трех массивах
arr1 = [3, 1, 4, 1, 5, 9, 2]
arr2 = [10, -5, 8, 0, -2]
arr3 = [7, 7, 7, 7]

print("Минимальные элементы:")
print(f"Массив 1: {MinRec(arr1, len(arr1))}")
print(f"Массив 2: {MinRec(arr2, len(arr2))}")
print(f"Массив 3: {MinRec(arr3, len(arr3))}")
```

```python
# 8) Функция Digits для подсчета цифр в строке
def Digits(S):
    if len(S) == 0:
        return 0
    else:
        first_char = S[0]
        rest_string = S[1:]
        if first_char.isdigit():
            return 1 + Digits(rest_string)
        else:
            return Digits(rest_string)

test_string = "Hello123World456"
print(f"Количество цифр в строке '{test_string}': {Digits(test_string)}")
```

```python
# 9) Функция Simm для проверки симметричности строки
def Simm(S):
    if len(S) <= 1:
        return True
    else:
        if S[0] == S[-1]:
            return Simm(S[1:-1])
        else:
            return False

test_string1 = "radar"
test_string2 = "hello"

print(f"Строка '{test_string1}' симметрична: {Simm(test_string1)}")
print(f"Строка '{test_string2}' симметрична: {Simm(test_string2)}")
```

```python
# 10) Функция для сложения двоичных чисел
def binary_add(a, b, carry=0):
    if a == '' and b == '':
        return str(carry) if carry else ''
    
    # Берем последние цифры или 0 если строка пустая
    bit_a = int(a[-1]) if a else 0
    bit_b = int(b[-1]) if b else 0
    
    # Вычисляем сумму битов и перенос
    total = bit_a + bit_b + carry
    current_bit = total % 2
    next_carry = total // 2
    
    # Рекурсивно обрабатываем остальные биты
    rest_a = a[:-1] if a else ''
    rest_b = b[:-1] if b else ''
    
    result = binary_add(rest_a, rest_b, next_carry)
    
    return result + str(current_bit)

# Тестируем сложение двоичных чисел
binary1 = "101"  # 5
binary2 = "1101" # 13
result = binary_add(binary1, binary2)
print(f"Сумма {binary1}(2) + {binary2}(2) = {result}(2)")
```

```python
# 11) Функция Root для нахождения корня методом деления отрезка пополам
def Root(f, a, b, epsilon):
    mid = (a + b) / 2
    if abs(b - a) < epsilon:
        return mid
    else:
        if f(a) * f(mid) < 0:
            return Root(f, a, mid, epsilon)
        else:
            return Root(f, mid, b, epsilon)

# Пример: находим корень уравнения x^2 - 4 = 0
def f(x):
    return x**2 - 4

root = Root(f, 0, 3, 0.0001)
print(f"Корень уравнения x^2 - 4 = 0 на [0, 3]: {root:.4f}")
```

```python
# 12) Функция для обращения числа без использования циклов и строк
def reverse_number(n, reversed_n=0):
    if n == 0:
        return reversed_n
    else:
        last_digit = n % 10
        remaining_number = n // 10
        new_reversed = reversed_n * 10 + last_digit
        return reverse_number(remaining_number, new_reversed)

number = 12345
reversed_num = reverse_number(number)
print(f"Число {number} в обратном порядке: {reversed_num}")
```

```python
# 13) Функция Simm для проверки симметричности части строки
def Simm_part(S, i, j):
    if i >= j:
        return True
    else:
        if S[i] == S[j]:
            return Simm_part(S, i + 1, j - 1)
        else:
            return False

test_string = "abcba"
print(f"Часть строки '{test_string}' с индексами 1-3 симметрична: {Simm_part(test_string, 1, 3)}")
```

```python
# 14) Функция для суммы цифр числа без использования циклов и строк
def sum_digits(n):
    if n == 0:
        return 0
    else:
        return n % 10 + sum_digits(n // 10)

number = 12345
print(f"Сумма цифр числа {number}: {sum_digits(number)}")
```

```python
# 15) Функция для проверки палиндрома без использования циклов и срезов
def is_palindrome(word, start=0, end=None):
    if end is None:
        end = len(word) - 1
    
    if start >= end:
        return "yes"
    else:
        if word[start] == word[end]:
            return is_palindrome(word, start + 1, end - 1)
        else:
            return "no"

test_word1 = "radar"
test_word2 = "hello"

print(f"Слово '{test_word1}' палиндром: {is_palindrome(test_word1)}")
print(f"Слово '{test_word2}' палиндром: {is_palindrome(test_word2)}")
```
