```python
def last_k_lines(K, input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    with open(output_file, 'w', encoding='utf-8') as f:
        f.writelines(lines[-K:])
```

```python
def remove_last_k_lines(K, filename):
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    with open(filename, 'w', encoding='utf-8') as f:
        f.writelines(lines[:-K])
```

```python
def remove_first_k_chars(K, filename):
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    with open(filename, 'w', encoding='utf-8') as f:
        for line in lines:
            f.write(line[K:] if len(line) > K else '')
```

```python
def append_lines(file1, file2):
    with open(file1, 'r', encoding='utf-8') as f:
        lines1 = f.readlines()
    with open(file2, 'r', encoding='utf-8') as f:
        lines2 = f.readlines()
    with open(file1, 'w', encoding='utf-8') as f:
        for i in range(len(lines1)):
            line1 = lines1[i].rstrip('\n')
            if i < len(lines2):
                line2 = lines2[i].rstrip('\n')
                f.write(f"{line1}{line2}\n")
            else:
                f.write(lines1[i])
```

```python
def remove_line_k(K, filename):
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    if 1 <= K <= len(lines):
        del lines[K-1]
        with open(filename, 'w', encoding='utf-8') as f:
            f.writelines(lines)
```

```python
import string

def extract_punctuation(input_file, output_file):
    punctuation_chars = []
    with open(input_file, 'r', encoding='utf-8') as f:
        for char in f.read():
            if char in string.punctuation:
                punctuation_chars.append(char)
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(''.join(punctuation_chars))
```

```python
def count_and_sum_numbers(filename):
    count = 0
    total = 0
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            number_str = line.strip()
            if number_str:  # проверка на пустую строку
                count += 1
                total += int(number_str)
    return count, total
```

```python
def replace_empty_lines(S, filename):
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    with open(filename, 'w', encoding='utf-8') as f:
        for line in lines:
            if line.strip() == '':
                f.write(f"{S}\n")
            else:
                f.write(line)
```

```python
def center_text(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    with open(filename, 'w', encoding='utf-8') as f:
        for line in lines:
            line = line.rstrip('\n')
            if line:
                line = line.rstrip()
                if len(line) % 2 == 1:
                    line = ' ' + line
                spaces = ' ' * ((50 - len(line)) // 2)
                f.write(f"{spaces}{line}\n")
            else:
                f.write('\n')
```

```python
def count_paragraphs(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        content = f.read()
    paragraphs = content.split('\n\n')
    count = 0
    for para in paragraphs:
        if para.strip():  # не считаем пустые абзацы
            count += 1
    return count
```

```python
def right_align_text(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    with open(filename, 'w', encoding='utf-8') as f:
        for line in lines:
            line = line.rstrip('\n')
            if line:
                line = line.rstrip()
                spaces = ' ' * (50 - len(line))
                f.write(f"{spaces}{line}\n")
            else:
                f.write('\n')
```

```python
def prepend_file(file1, file2):
    with open(file2, 'r', encoding='utf-8') as f:
        content2 = f.read()
    with open(file1, 'r', encoding='utf-8') as f:
        content1 = f.read()
    with open(file1, 'w', encoding='utf-8') as f:
        f.write(content2 + content1)
```

```python
def insert_empty_line_before_k(K, filename):
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    if 1 <= K <= len(lines):
        lines.insert(K-1, '\n')
        with open(filename, 'w', encoding='utf-8') as f:
            f.writelines(lines)
```

```python
def duplicate_empty_lines(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    result = []
    for line in lines:
        result.append(line)
        if line.strip() == '':
            result.append(line)
    with open(filename, 'w', encoding='utf-8') as f:
        f.writelines(result)
```

```python
def remove_first_last_lines(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    if len(lines) >= 2:
        lines = lines[1:-1]
        with open(filename, 'w', encoding='utf-8') as f:
            f.writelines(lines)
```
