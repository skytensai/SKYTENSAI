```python
#1
def Mean(X, Y):
    AMean = (X + Y) / 2
    GMean = (X * Y) ** 0.5
    return AMean, GMean

X = float(input("X: "))
Y = float(input("Y: "))
a, g = Mean(X, Y)
print(f"Среднее арифметическое: {a}")
print(f"Среднее геометрическое: {g}")
```

```python
#2
import math

def Circle(R):
    area = math.pi * R ** 2
    circumference = 2 * math.pi * R
    return area, circumference

R = float(input("Радиус: "))
area, circ = Circle(R)
print(f"Площадь: {area:.2f}")
print(f"Длина окружности: {circ:.2f}")
```

```python
#3
import math

def TrianglePS(a):
    P = 3 * a
    S = (math.sqrt(3) / 4) * a ** 2
    return P, S

a = float(input("Сторона треугольника: "))
P, S = TrianglePS(a)
print(f"Периметр: {P}")
print(f"Площадь: {S:.2f}")
```

```python
#4
import math

def RingS(R1, R2):
    S1 = math.pi * R1 ** 2
    S2 = math.pi * R2 ** 2
    return abs(S1 - S2)

R1 = float(input("R1: "))
R2 = float(input("R2: "))
area = RingS(R1, R2)
print(f"Площадь кольца: {area:.2f}")
```

```python
#5
def RectPS(x1, y1, x2, y2):
    width = abs(x2 - x1)
    height = abs(y2 - y1)
    P = 2 * (width + height)
    S = width * height
    return P, S

x1, y1 = map(float, input("x1 y1: ").split())
x2, y2 = map(float, input("x2 y2: ").split())
P, S = RectPS(x1, y1, x2, y2)
print(f"Периметр: {P}")
print(f"Площадь: {S}")
```

```python
#6
import math

def TriangleP(a, h):
    b = math.sqrt((a/2)**2 + h**2)
    P = a + 2 * b
    return P

a = float(input("Основание: "))
h = float(input("Высота: "))
P = TriangleP(a, h)
print(f"Периметр: {P:.2f}")
```

```python
#7
def InvertDigits(K):
    return int(str(K)[::-1])

K = int(input("Число: "))
inverted = InvertDigits(K)
print(f"Обратное число: {inverted}")
```

```python
#8
def SumRange(A, B):
    if A > B:
        return 0
    return sum(range(A, B + 1))

A = int(input("A: "))
B = int(input("B: "))
result = SumRange(A, B)
print(f"Сумма: {result}")
```

```python
#9
import math

def TypeTr1(x1, y1, x2, y2, x3, y3):
    def distance(x1, y1, x2, y2):
        return math.sqrt((x2-x1)**2 + (y2-y1)**2)
    
    a = distance(x1, y1, x2, y2)
    b = distance(x2, y2, x3, y3)
    c = distance(x3, y3, x1, y1)
    
    if a == b == c:
        return "Равносторонний"
    elif a == b or b == c or a == c:
        return "Равнобедренный"
    elif abs(a**2 + b**2 - c**2) < 0.001 or abs(a**2 + c**2 - b**2) < 0.001 or abs(b**2 + c**2 - a**2) < 0.001:
        return "Прямоугольный"
    else:
        return "Обычный"

x1, y1 = map(float, input("x1 y1: ").split())
x2, y2 = map(float, input("x2 y2: ").split())
x3, y3 = map(float, input("x3 y3: ").split())
print("Треугольник:", TypeTr1(x1, y1, x2, y2, x3, y3))
```

```python
#10
def Quarter(x, y):
    if x > 0 and y > 0:
        return 1
    elif x < 0 and y > 0:
        return 2
    elif x < 0 and y < 0:
        return 3
    elif x > 0 and y < 0:
        return 4

x = float(input("x: "))
y = float(input("y: "))
print(f"Четверть: {Quarter(x, y)}")
```

```python
#11
def DigitCountSum(K):
    digits = str(K)
    count = len(digits)
    total = sum(int(d) for d in digits)
    return count, total

K = int(input("K: "))
count, total = DigitCountSum(K)
print(f"Количество цифр: {count}")
print(f"Сумма цифр: {total}")
```

```python
#12
def Calc(A, B, Op):
    if Op == 1:
        return A - B
    elif Op == 2:
        return A * B
    elif Op == 3:
        return A / B
    else:
        return A + B

A = float(input("A: "))
B = float(input("B: "))
Op = int(input("Op (1-вычитание, 2-умножение, 3-деление): "))
result = Calc(A, B, Op)
print(f"Результат: {result}")
```

```python
#13
def LuckyNum(num):
    digits = [int(d) for d in str(num)]
    return sum(digits[:2]) == sum(digits[2:])

print("Счастливые номера:")
for num in range(1000, 10000):
    if LuckyNum(num):
        print(num, end=" ")
```

```python
#14
import math

def DegToRad(D):
    return D * math.pi / 180

D = float(input("Градусы: "))
radians = DegToRad(D)
print(f"Радианы: {radians:.4f}")
```

```python
#15
def IsLeapYear(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

year = int(input("Год: "))
print("Високосный" if IsLeapYear(year) else "Не високосный")
```
